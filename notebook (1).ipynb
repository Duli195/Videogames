```python
from scipy import stats
xbox = df[(df['platform'] == 'XOne') & (df['user_score'].notna())]['user_score']
pc = df[(df['platform'] == 'PC') & (df['user_score'].notna())]['user_score']
alpha = 0.05
t_stat, p_val = stats.ttest_ind(xbox, pc, equal_var=False)

print('Estadístico t:', t_stat)
print('Valor p:', p_val)

if p_val < alpha:
    print('Se rechaza la hipótesis nula: las calificaciones promedio SON diferentes.')
else:
    print('No se puede rechazar la hipótesis nula: NO hay diferencia significativa.')
```

```python
accion = df[(df['genre'] == 'Action') & (df['user_score'].notna())]['user_score']
deportes = df[(df['genre'] == 'Sports') & (df['user_score'].notna())]['user_score']
t_stat2, p_val2 = stats.ttest_ind(accion, deportes, equal_var=False)

print('Estadístico t:', t_stat2)
print('Valor p:', p_val2)

if p_val2 < alpha:
    print('Se rechaza la hipótesis nula: las calificaciones promedio SON diferentes.')
else:
    print('No se puede rechazar la hipótesis nula: NO hay diferencia significativa.')
```

```python
from scipy import stats
import pandas as pd
df = pd.read_csv('/datasets/games.csv')
print(df.columns)
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')
xbox = df[(df['Platform'] == 'XOne') & (df['User_Score'].notna())]['User_Score']
pc = df[(df['Platform'] == 'PC') & (df['User_Score'].notna())]['User_Score']
alpha = 0.05
levene_stat, levene_p = stats.levene(xbox, pc)
print('Estadístico de Levene:', levene_stat)
print('Valor p de Levene:', levene_p)
equal_var = True if levene_p > 0.05 else False
t_stat, p_val = stats.ttest_ind(xbox, pc, equal_var=equal_var)
print('Estadístico t:', t_stat)
print('Valor p:', p_val)
if p_val < 0.05:
    print('Se rechaza la hipótesis nula: las calificaciones promedio SON diferentes.')
else:
    print('No se puede rechazar la hipótesis nula: NO hay diferencia significativa.')
```

Adoptaremos un nivel de significancia de 0.05; si el valor p resultante es mayor a 0.05, no podemos rechazar la hipótesis nula, y si es menor a 0.05, rechazamos la hipótesis nula, indicando que las varianzas no son iguales.
